```python
#This method generates a report of missing values in any dataframe
def df_missing_report_generator(df):
    missing=len(df)-df.count()
    missing_perct=(((missing)*100)/len(df)).astype(str)
    missing_perct=missing_perct.apply(lambda x: x+' %')
    Missing_value_report=pd.concat([missing,missing_perct], axis='columns')
    Missing_value_report.rename(columns={0:'Missing_values',1:'Percentage_Missing_values'},inplace=True)
    return Missing_value_report
```

```python
#This method generates a list of outliers of a data column(or series data)
def detect_outliers(data):
    outliers=[]
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

```python
# The Release Clause data column has no outliers, hence the outlier list is empty
detect_outliers(processing_df['Release Clause'])
```

```python
# This boxplot also supports above conclusion and shows no outliers
sns.boxplot(processing_df['Release Clause'])
```

```python
#This loop fills the missing values of columns with median if the column data has outliers else with mean value 
for col in numeric_col_lst:
    if detect_outliers(processing_df[col]):
        processing_df[col].fillna(processing_df[col].median(),inplace=True)
    else:
        processing_df[col].fillna(processing_df[col].mean(),inplace=True)
```

```python
#This report shows that the missing values of numeric data columns are also filled
df_missing_report_generator(processing_df[numeric_col_lst])
```
